# Mathematical Operators

**Introduction**

Mathematical operators can broadly be classified into;

- Arithmetic operators

- Assignment operators

- Comparison operators

- Logical operators

- Identity operators

- Bitwise operators

**Arithmetic Operations**

```python
# Addition
print("Addition:", 5 + 3)
```

```python
# Subtraction
print("Subtraction:", 10 - 3)
```

```python
# Multiplication
print("Multiplication:", 4 * 5)
```

```python
# Division
print("Division:", 20 / 4)
```

```python
# Incorrect usage: Adding an integer and a string
print(5 + "3")
```

**Assignment operators**

```python
# Assignment
x = 10
print("Assignment:", x)
```

```python
# Increment
x += 5
print("Increment:", x)
```

```python
# Decrement
x -= 3
print("Decrement:", x)
```

```python
# Multiply and assign
x *= 2
print("Multiply and Assign:", x)
```

```python
# Correct: dividing an integer by an integer
print(x / int("124"))
```

**Comparison operators**

```python
# Comparison
a = 15
b = 10

# Greater than
print("Greater than:", a > b)
```

```python
# Less than or equal to
print("Less than or equal to:", a <= b)
```

```python
# Equal to
print("Equal to:", a == b)
```

```python
# Not equal to
print("Not equal to:", a != b)
```

```python
# Incorrect usage: attempting to compare a string and a number
print("2" > 3)
```

```python
# Corrected: compare a number and another number
print(2 > 3)
```

**Logical operators**

```python
# Logical
statement1 = True
statement2 = False

# AND
print("Result of AND:", (statement1 and statement2))
```

```python
# OR
print("Result of OR:", (statement1 or statement2))
```

```python
# NOT
print("Result of NOT:", (not statement1))
```

```python
# Combining conditions with AND and OR
age = 25
is_student = False

# Both conditions must be True
if age > 18 and not is_student:
    print("You are an adult, not a student.")
```

```python
# Incorrect usage: Using AND operator with non-boolean values
print("Incorrect Result with AND operator:", ("Hello" and 42))
```

```python
# Corrected using boolan values and logical AND
print("Corrected result:",True and False)
```

**Bitwise Operators**

Bitwise operators are operators that perform operations directly to the binary representation of Integers

```python

a = 5      # Binary: 0101
b = 3      # Binary: 0011

# AND
print("Bitwise AND:", (a & b))
```

```python
# OR
print("Bitwise OR:", (a | b))

```

```python
 # XOR
print("Bitwise XOR:", (a ^ b))

```

```python
# NOT
print("Bitwise NOT:",~a)
```

**Identity Operators**

They are used to compare the **memory location** of two objects

*Identity operators are used to check whether two variables refer to the exact same object in memory, not just whether they have the same value.*

```python
# Identity operators with numbers
a = 10
b = 10.0

print("is:", a is b) #False sincce a and b are diffrent types
```

```python
print("is not:", a is not b) #True, as a and b are different types
```
